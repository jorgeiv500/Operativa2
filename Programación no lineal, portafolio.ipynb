<div style="text-align:justify;">
En la actualidad, una práctica común entre los administradores de grandes carteras de inversión es usar, como guía, modelos de computadora basados en parte en programación no lineal. Debido a que los inversionistas se preocupan tanto por el <i><b>rendimiento esperado (ganancia)</b></i>, como por el <i><b>riesgo asociado con su inversión</b></i>, la programación no lineal se emplea para determinar una cartera que, bajo ciertos supuestos, proporcione un trueque óptimo entre estos dos factores. Este enfoque se basa en gran parte en la novedosa investigación realizada por <i><b>Harry Markowitz y William Sharpe</b></i>, que los ayudó a ganar el Premio Nobel de Economía en 1990.
</div>

```julia
using JuMP,Ipopt
covar=[0.001600808	0.001133842	0.000593164
0.001133842	0.003402472	0.000549621
0.000593164	0.000549621	0.000664277]
rend=[-0.001313115	-0.003464184	0.001669026]
precio=[10.7	1.04	286.920013]
acciones = Model(Ipopt.Optimizer)
@variable(acciones,x[1:1,1:3]>=0)
@objective(acciones,Min,sum((x*covar).*x))
@constraint(acciones,sum(x)==1)
#@constraint(acciones,sum(rend.*x)>=0.001)
#@constraint(acciones,sum(precio.*x)<=2000)
print(acciones)
JuMP.optimize!(acciones)
println("El minimo riesgo es",JuMP.getobjectivevalue(acciones))
println("las asignaciones son", JuMP.value.(x))
```

```julia
from pyomo.environ import *
model = ConcreteModel()
model.x = Var(domain=NonNegativeReals,initialize=10) #,initialize=15
model.objective = Objective(expr = 12*model.x**5-975*model.x**4+28000*model.x**3-345000*model.x**2+1800000*model.x, sense=maximize)
model.demand = Constraint(expr = model.x <= 31)
results = SolverFactory('ipopt').solve(model) 
print('x es igual a = ', model.x())
print('\nOptimo =', model.objective())
```
