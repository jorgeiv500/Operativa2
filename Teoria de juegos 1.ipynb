## FORMULACI√ìN DE JUEGOS DE DOS PERSONAS Y SUMA CERO

Para ilustrar las caracter√≠sticas b√°sicas de un modelo de juegos de dos personas de suma cero, considere el juego llamado **pares ‚úåÔ∏è y nones üëÜ**. √âste consiste nada m√°s en que los dos jugadores muestran al mismo tiempo uno o dos dedos. 

> Si el n√∫mero total de dedos mostrados por ambos jugadores es par ‚úåÔ∏è, el jugador que apuesta a pares (por ejemplo, el jugador 1) gana la apuesta (digamos üíµ1 d√≥larüíµ) al jugador que elige nones (jugador 2). 

> Si el n√∫mero de dedos es imparüëÜ, **el jugador 1**ü§®üôÅ paga üíµ1 d√≥larüíµ al **jugador 2**üòÑü§ë. Entonces, cada jugador tiene dos estrategias: mostrar uno o dos dedos. El pago en d√≥lares que resulta para el jugador 1 se muestra en una matriz de pagos en la siguiente tabla.

<img src="https://github.com/jorgeiv500/Investigacion-Operativa/blob/master/est.png?raw=true" style="height: 12em; width:auto" >

En general, un juego de dos personas se caracteriza por

1. Las estrategias del jugador 1.
2. Las estrategias del jugador 2.
3. La matriz de pagos.

> * Antes de iniciar el juego, cada jugador conoce las estrategias de que dispone, las que tiene su oponente y la matriz de pagos. 
* Una jugada real consiste en que los dos jugadores elijan al mismo tiempo una estrategia sin saber cu√°l es la elecci√≥n de su oponente.
* Una estrategia puede constituir una acci√≥n sencilla, como mostrar un n√∫mero par o non de dedos en el juego de pares y nones. 

Por otro lado, en juegos üé∞üé∞ m√°s complicados que llevan en s√≠ una serie de movimientos, una estrategia es una regla predeterminada que especifica por completo c√≥mo se intenta responder a cada circunstancia posible en cada etapa del juego. 

____________________

Por ejemplo, una estrategia de un jugador de ajedrez‚ôüÔ∏è‚ôüÔ∏è‚ôüÔ∏è indica c√≥mo hacer el siguiente movimiento ante todas las posiciones posibles en el tablero, por lo que el n√∫mero total de estrategias posibles ser√≠a astron√≥mico.

_______________________

* Las aplicaciones de la teor√≠a de juegos involucran situaciones competitivas mucho menos complicadas que el ajedrez, pero las estrategias que se manejan pueden llegar a ser bastante complejas.

* Por lo general, la **matriz de pagos** muestra la ganancia (positiva o negativa) del jugador 1, que resultar√≠a con cada combinaci√≥n de estrategias de los dos jugadores. Se presenta s√≥lo la matriz del jugador 1, puesto que la del jugador 2 es el negativo de √©sta, debido a la naturaleza de suma cero del juego.


* Los elementos de la matriz de pagos pueden expresar cualquier tipo de unidades, como d√≥lares, siempre que representen con exactitud la utilidad del jugador 1 ante el resultado correspondiente.

* Debe hacerse hincapi√© en que la utilidad no necesariamente es proporcional a la cantidad de dinero (o cualquier otro bien) cuando se manejan cantidades grandes. 

> Por ejemplo, 2 millones de d√≥lares (despu√©s de impuestos) pueden tener un valor mucho menor que ‚Äúel doble‚Äù del valor que representa 1 mill√≥n de d√≥lares para una persona pobre. 

* En otras palabras, si a una persona se le da a elegir entre: 

> 1. recibir, con 50% de posibilidades, 2 millones o nada y 
> 2. recibir 1 mill√≥n con seguridad, una persona pobre tal vez preferir√≠a la √∫ltima oferta. 

* En otras palabras, el resultado que corresponde a un elemento con valor 2 en una matriz de pagos debe ‚Äúvaler el doble‚Äù para el jugador 1 que el resultado correspondiente a un elemento de 1. As√≠, dada la elecci√≥n, debe serle indiferente 50% de posibilidades de recibir el primer resultado (en lugar de nada) y recibir en definitiva el √∫ltimo resultado.


* Un objetivo primordial de la teor√≠a de juegos es desarrollar criterios racionales para seleccionar una estrategia, los cuales implican dos supuestos importantes:

> 1. Ambos jugadores son racionales.
> 2. Ambos jugadores eligen sus estrategias s√≥lo para promover su propio bienestar (sin compasi√≥n para el oponente).

* La teor√≠a de juegos se contrapone al an√°lisis de decisi√≥n, en donde se hace el supuesto de que el tomador de decisiones est√° jugando un juego contra un oponente pasivo ‚Äîla naturaleza‚Äî que elige sus estrategias de alguna manera aleatoria.

### Variaci√≥n 2 del ejemplo

Ahora suponga que los datos actuales se dan en la **tabla 3** como la matriz de pagos del jugador 1 (pol√≠tico 1). Este juego no tiene estrategias dominadas por lo que no es obvio qu√© deben hacer los jugadores. ¬øQu√© l√≠nea de razonamiento dice la teor√≠a de juegos que debe usarse en este caso?

* **Considere al jugador 1.** Si elige la estrategia 1 puede ganar 6 o perder 3. Como el jugador 2 es racional y tratar√° de aplicar una estrategia que lo proteja de pagos grandes al jugador 1, parece probable que si juega la estrategia 1, el jugador 1 perder√°. De manera similar, si selecciona la estrategia 3 el jugador 1 puede ganar 5, pero quiz√° sea m√°s probable que su oponente racional evite esta p√©rdida y en su lugar logre que √©l pierda, lo que puede ascender a 4. Por otro lado, si el jugador 1 elige la estrategia 2, tiene garantizado que no perder√° y quiz√° gane algo. Entonces, por proporcionar la mejor garant√≠a (un pago de 0), la estrategia 2 parece ser la elecci√≥n ‚Äúracional‚Äù del jugador 1 contra su oponente racional. (Esta l√≠nea de razonamiento supone que ambos jugadores tienen aversi√≥n a arriesgar p√©rdidas m√°s grandes que las necesarias, al contrario de quienes disfrutan la apuesta por un gran pago con pocas posibilidades.)

* **Ahora considere al jugador 2.** Puede perder tanto como 5 o 6 si sigue las estrategias 1 o 3, pero est√° garantizado que al menos empata con la estrategia 2. Entonces, si usa el mismo razonamiento para buscar su mejor garant√≠a contra su oponente racional, parece que la mejor elecci√≥n es la estrategia 2.

Si los dos jugadores eligen la estrategia 2, el resultado es que empatan. As√≠, en este caso, ning√∫n jugador mejora con su mejor garant√≠a, pero ambos fuerzan al oponente hacia la misma posici√≥n. Aunque cada jugador deduzca la estrategia del otro, no puede explotar esta informaci√≥n para mejorar su posici√≥n. Se trata de un empate.

El producto final de esta l√≠nea de razonamiento es que cada jugador debe jugar de tal manera que minimice su p√©rdida m√°xima, siempre que el resultado de su elecci√≥n no pueda ser aprovechado por su oponente para mejorar su posici√≥n. Esto se conoce como criterio minimax, criterio est√°ndar que propone la teor√≠a de juegos para elegir una estrategia. En realidad, este criterio sostiene que se debe seleccionar la mejor estrategia, aun cuando la elecci√≥n fuera anunciada al oponente antes de que √©ste eligiera su estrategia. En t√©rminos de la matriz de pagos, implica que el jugador 1 debe elegir aquella estrategia cuyo pago m√≠nimo sea el mayor, mientras que el jugador 2 debe elegir aquella cuyo pago m√°ximo al jugador 1 sea el menor. Este criterio se muestra en la **tabla 3**, donde se identifi ca la estrategia 2 como la estrategia maximin del jugador 1, y la estrategia 2 es la estrategia minimax del jugador 2. El pago de 0 que resulta es el valor del juego, por lo que √©ste es un juego justo.

```python
import numpy as np
import matplotlib.pyplot as plt


# Construct lines
# x > 0
x = np.linspace(0, 20, 2000)
# y >= 2
y1 = (x*0) + 2
# 2y <= 25 - x
y2 = (25-x)/2.0
# 4y >= 2x - 8 
y3 = (2*x-8)/4.0
# y <= 2x - 5 
y4 = 2 * x -5

# Make plot
plt.plot(x, y1, label=r'$y\geq2$')
plt.plot(x, y2, label=r'$2y\leq25-x$')
plt.plot(x, y3, label=r'$4y\geq 2x - 8$')
plt.plot(x, y4, label=r'$y\leq 2x-5$')
plt.xlim((0, 16))
plt.ylim((0, 11))
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')

# Fill feasible region
y5 = np.minimum(y2, y4)
y6 = np.maximum(y1, y3)
plt.fill_between(x, y5, y6, where=y5>y6, color='grey', alpha=0.5)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
